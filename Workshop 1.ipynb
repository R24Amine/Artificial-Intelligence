```python
#REMPLACER LES ESPACES DANS LES NOMS DES COLONNES PAR DES ' _ '
Mydataframe.columns = Mydataframe.columns.str.replace(' ', '_')
Mydataframe.head()
#Mydataframe.columns = df.columns.str.lstrip()
#Mydataframe.columns = df.columns.str.rstrip()
#Mydataframe.columns.str.strip()
#To replace white space at the beginning: Mydataframe.columns = Mydataframe.columns.str.replace('^ +', '_')
#To replace white space at the end: Mydataframe.columns = Mydataframe.columns.str.replace(' +$', '_')
#To replace white space at both ends: Mydataframe.columns = Mydataframe.columns.str.replace('^ +| +$', '_')


```

```python
#Modifier le dataframe original en remplacant les valeurs manquantes par des zeros
Mydataframe.fillna(0,inplace=True)
Mydataframe.head()
```

```python
#creer une nouvelle colonne qui est la somme d'autres colonnes
Mydataframe['Books_appearances'] = Mydataframe['GoT']+  Mydataframe['CoK'] +  Mydataframe['SoS'] +  Mydataframe['FfC'] + Mydataframe['DwD']
Mydataframe.head()
```

```python
#utilisation de la clause where pour modifier les valeurs d'une colonne
Mydataframe['Gender'] = np.where(Mydataframe['Gender']==0, 'F', 'M')
```

```python
#creation d'une nouvelle colonne en utilisant where
Mydataframe['AllBooks'] = np.where(Mydataframe['Books_appearances']==5, True, False)
```

```python
#affichage avec condition sur les colonnnes
Mydataframe[(Mydataframe['AllBooks']==True) & (Mydataframe['Death_Year'] ==0)]
```

```python
#Boule for pour créer une nouvelle colonne en appliquant des conditions sur plusiseurs colonnes
for index, row in Mydataframe.iterrows():
    
    if (row['GoT']==1):
        reponse = "GoT"
    elif (row['GoT']==0) and (row['CoK']==1):
        reponse = "Cok"
    elif (row['GoT']==0) and (row['CoK']==0) and (row['SoS']==1):
        reponse = "SoS"
    elif (row['GoT']==0) and (row['CoK']==0) and (row['SoS']==0) and (row['FfC']==1):
        reponse = "Ffc"
    elif (row['GoT']==0) and (row['CoK']==0) and (row['SoS']==0) and (row['FfC']==0) and (row['DwD']==1):
        reponse = "Dwd"
    Mydataframe.at[index,'First_Book_Appearence'] = reponse
Mydataframe.head()
```

```python

Type_count  = Mydataframe['First_Book_Appearence'].value_counts()
plt.figure(figsize=(12,7))
plt.bar(Type_count.index, Type_count.values, alpha=0.9,color = ['red','green','blue','grey','yellow'])
plt.title('Count of characters introduced by books')
plt.ylabel('Number of characters', fontsize=12)
plt.xlabel('first appearance', fontsize=12)
plt.show()
```

```python
Type_count  = Mydataframe['First_Book_Appearence'].value_counts()
plt.figure(figsize=(12,7))
plt.barh(Type_count.index, Type_count.values, alpha=0.9,color = ['red','green','blue','grey','yellow'])
plt.title('Count of characters introduced by books')
plt.ylabel('Number of characters', fontsize=12)
plt.xlabel('first appearance', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(4,3))
Type_count  = Mydataframe['First_Book_Appearence'].value_counts()
plt.pie(Type_count.values, labels=Type_count.index, startangle=90, autopct='%1.1f%%',shadow=True,radius=2.7 ,colors = ['red','green','blue','grey','yellow'])
plt.show()
```
